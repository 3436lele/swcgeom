```python
import sys
from functools import reduce
from operator import add
import matplotlib.pyplot as plt

import swcgeom
from swcgeom.analysis import draw
from swcgeom.transforms import TreeNormalizer
```

```python
swc1 = "101711-10_4p5-of-16_initial.CNG.swc"
swc2 = "101711-11_16-of-16_initial.CNG.swc"
```

```python
# create a tree
tree = swcgeom.Tree.from_swc(swc1)

# print tree
print("print tree:", tree)
print("print node:", tree[1])
for a in tree[:3]: print(a)

# generate swc file
# tree.to_swc("./test.swc")

segments = tree.get_segments()
eq = len(segments) == tree.number_of_edges()
print(f"Neuron tree with {len(segments)} segments,", eq)

branches = tree.get_branches()
print(f"Neuron tree with {len(branches)} branches.")

# collect tips
tips = tree.traverse(leave=lambda cur, children: max(1, reduce(add, children, 0)))
print(f"tips: {tips}")

# darw
ax, lines = draw(tree)
```

```python
normalizer = TreeNormalizer()
normalized_tree = normalizer(tree)
_, _ = draw(normalized_tree)
```

```python
fig, ax = plt.subplots(1,1)
_ = draw(swcgeom.Tree.from_swc(swc1), ax=ax, color="red")
_ = draw(swcgeom.Tree.from_swc(swc2), ax=ax, color="green", first=False)
```

```python
def cut_3(n: swcgeom.Tree.Node, _) -> tuple[None, bool]:
    return None, n.id == 3
tree_cut3 = swcgeom.tree_utils.cut_tree(tree, leave=cut_3)

fig, (ax0, ax1) = plt.subplots(1, 2, sharex=True, sharey=True)
draw(tree, ax=ax0)
draw(tree_cut3, ax=ax1)

def cut_level(max_level: int):
    def cut(n: swcgeom.Tree.Node, parent_level: int | None) -> tuple[int, bool]:
        level = 0 if parent_level is None else parent_level + 1
        return (level, level > max_level)
    return cut

fig, axes = plt.subplots(2, 5, sharex=True, sharey=True)
axes = axes.flatten()
for i in range(10):
    tree_cut100 = swcgeom.tree_utils.cut_tree(tree, enter=cut_level(i+3))
    draw(tree_cut100, ax=axes[i])
```
